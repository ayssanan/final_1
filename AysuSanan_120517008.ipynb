The Theoretical Part

I set out from the formula F_net=0 to study the energy diagram of an electron in a changing electric field.
In my question, the force F was equal to the formula = q(E+V*B).
I have arranged this formula as follows.

m**a** = q(**E** +**v** x **B**)

**a** = q\m(**E** +**V** x **B** )

In order to do this analysis and to convert the analysis into code, we need 2 correlations.These are the general correlations given below ;

dx/dt = v (The first derivative of position by time gives the speed.)

d^2x^2/ dt^2 = a (The second derivative of position with respect to time gives the acceleration.)

In my position, I used these general correlations as follows.

v=a*dt

**F**=v*dt

I used the matrix method when doing my analysis and converting this analysis into code.Thanks to this method, I have simplified the analysis part, which is more complicated with unit vectors.Instead of unit vectors, I obtained different states by changing the **B,V,E** magnitudes inside the matrix.I have put these changes on the graph, detailed the graphs, and detailed the necessary analyzes.





Note:Bold means Vector Quantity.

```python
import numpy as np
from matplotlib import pyplot as plt
q = 1
m = 1
v = np.array([0,0,0])
E = np.array([1,0,0])
B = np.array([0,1,0])
r = np.array([0,0,0])
a = 0
dt = 0.001
t = 0
tlist = [0]
xlist = [r[0]]
ylist = [r[1]]
zlist = [r[2]]
for i in np.arange(10/dt): # i used 10\dt because I saw the same amount of effect on changing dt.
    dv = (q/m)*np.add(E,(np.cross(v,B))) * dt
    v = np.add(v,dv)
    r = np.add(r,v*dt)
    t += dt
    tlist.append(t)
    xlist.append(r[0])
    ylist.append(r[1])
    zlist.append(r[2])
    
    print(r)

plt.plot(tlist,xlist)
plt.plot(tlist,ylist)
plt.plot(tlist,zlist)
plt.show()
```
